```python
def convert_dollar_to_int(value):
    if isinstance(value, str):
        value = value.replace('$', '')  # Remove "$" symbol
        if value.endswith('B'):
            return int(value[:-1]) * 1000000000  # Convert billion to integer
        elif value.endswith('M'):
            return int(value[:-1]) * 1000000  # Convert million to integer
        else:
            return int(value)  # Convert other numeric values to integer
    return value  # Return NaN values as they are

# Apply the conversion function to the 'Value' column
df['Funding'] = df['Funding'].apply(convert_dollar_to_int)

# Calculate the median of the integer values
median_value = df['Funding'].median()

# Replace NaN values with the median value
df['Funding'].fillna(median_value, inplace=True)
```
